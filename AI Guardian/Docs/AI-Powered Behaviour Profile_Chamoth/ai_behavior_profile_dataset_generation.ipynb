```python
import random
import uuid
import datetime
import json

# Random seed for reproducibility
random.seed(42)

# Function to generate a single patient record
def generate_patient_record(patient_id):
    age = random.randint(60, 90)
    gender = random.choice(["Male", "Female"])

    # Observation timestamps
    start_time = datetime.datetime.now()
    end_time = start_time + datetime.timedelta(hours=random.randint(6, 24))

    # Nursing notes
    nursing_notes = [
        "Patient refused morning medication citing dizziness.",
        "Noted increased agitation during evening rounds.",
        "Slept well through the night. Mood appeared improved.",
        "Observed to skip lunch and reported feeling fatigued."
    ]
    nursing_note = random.choice(nursing_notes)

    # Medications
    medications = [
        {"name": "Atorvastatin", "dose": f"{random.choice([25, 50, 75, 100])}mg",
         "scheduledTime": str(datetime.datetime.now()), "complianceStatus": random.choice(["Taken", "Missed", "Delayed"])},
        {"name": "Lisinopril", "dose": f"{random.choice([25, 50, 75, 100])}mg",
         "scheduledTime": str(datetime.datetime.now()), "complianceStatus": random.choice(["Taken", "Missed", "Delayed"])},
        {"name": "Paracetamol", "dose": f"{random.choice([25, 50, 75, 100])}mg",
         "scheduledTime": str(datetime.datetime.now()), "complianceStatus": random.choice(["Taken", "Missed", "Delayed"])}
    ]
    selected_medications = random.sample(medications, k=random.randint(1, 3))

    # Vitals
    vitals = {
        "heartRate": {"value": random.randint(60, 100), "unit": "bpm"},
        "spo2": {"value": random.randint(90, 100), "unit": "%"},
        "temperature": {"value": round(random.uniform(36.0, 39.0), 1), "unit": "C"},
        "bloodPressure": {"systolic": random.randint(100, 140), "diastolic": random.randint(60, 90), "unit": "mmHg"}
    }

    # ADLs
    adls = {
        "stepsTaken": random.randint(500, 8000),
        "calorieIntake": random.randint(1200, 2500),
        "sleepHours": round(random.uniform(4.0, 9.0), 1),
        "waterIntakeMl": random.randint(500, 2500),
        "mealsSkipped": random.randint(0, 2),
        "exerciseMinutes": random.randint(0, 60)
    }

    # Behaviour and emotion tags
    behaviour_tags = random.sample(["agitated", "withdrawn", "refusedActivity", "nonCompliant"], k=random.randint(0, 2))
    emotion_tags = random.sample(["calm", "anxious", "sad", "happy", "frustrated"], k=random.randint(0, 2))

    # Clinical summary
    clinical_summary = f"Patient status: {nursing_note} Continue monitoring."

    # Combine into record
    record = {
        "patientId": patient_id,
        "age": age,
        "gender": gender,
        "observationStart": str(start_time),
        "observationEnd": str(end_time),
        "nursingNote": nursing_note,
        "medications": selected_medications,
        "vitals": vitals,
        "adls": adls,
        "behaviourTags": behaviour_tags,
        "emotionAnalysis": {"tags": emotion_tags},
        "clinicalSummary": clinical_summary
    }

    return record
```

```python
# Generate dataset for 50 patients
dataset = [generate_patient_record(f"P{str(i).zfill(4)}") for i in range(1, 51)]

# Save to JSON
with open("ai_behavior_profile_dataset.json", "w") as f:
    json.dump(dataset, f, indent=4)

# Convert to CSV (flatten nested structure)
import pandas as pd
df = pd.json_normalize(dataset)
df.to_csv("ai_behavior_profile_dataset.csv", index=False)
```
