```python
# =========================================================
# Guardian Alerts (Sprint 2) â€” Daily Aggregation + Robust Cleaning (FINAL)
# Outputs: user_id, timestamp, anom_score, clf_prob, risk_level, reason
# =========================================================

# --- Light deps install (only if missing) ---
try:
    import torch, pandas, sklearn, matplotlib
except Exception:
    import sys
    !{sys.executable} -m pip install --quiet torch scikit-learn pandas numpy joblib matplotlib

# --- Imports / config ---
import os, glob, re, json, math, warnings
from typing import List, Tuple
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import IsolationForest, RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import joblib

warnings.filterwarnings("ignore")

try:
    import torch
    from torch import nn
    TORCH_OK = True
except Exception:
    TORCH_OK = False
    torch = None
    nn = None

# -------------------------
# Tunables
# -------------------------
ENGINE = "lstm"      # "lstm" or "iforest" (falls back to iforest if torch unavailable)
CLF    = "rf"        # "rf" or "mlp"
SEQ_LEN = 7         # days (post daily-aggregation)
TEST_SIZE = 0.2
RANDOM_STATE = 42

ID_COL = "user_id"
TS_COL = "timestamp"

BEHAV = ["steps","calorie_intake","sleep_hours","water_intake","bathroom_visits","exercise_minutes"]
VITAL = ["heart_rate","spo2","temperature","bp_sys","bp_dia","meals_skipped"]
FULL_FEATURES = BEHAV + VITAL

OUTDIR = "artifacts"
os.makedirs(OUTDIR, exist_ok=True)

# -------------------------
# Helpers: dataset discovery (same folder)
# -------------------------
def find_dataset_here() -> Tuple[str, str]:
    cands = []
    for ext in ("*.csv","*.xlsx","*.xls"):
        cands += glob.glob(os.path.join(".", ext))
    if not cands:
        # also look one level down just in case
        for ext in ("*.csv","*.xlsx","*.xls"):
            cands += glob.glob(os.path.join(".", "*", ext))
    if not cands:
        raise FileNotFoundError("No CSV/Excel found in current directory.")
    def score(p):
        n = os.path.basename(p).lower()
        s = 0
        if "new ai spreadsheet" in n: s += 5
        if "sheet1" in n: s += 2
        if n.endswith(".csv"): s += 1
        return -s
    cands.sort(key=score)
    chosen = cands[0]
    kind = "csv" if chosen.lower().endswith(".csv") else "excel"
    print(f"[data] Detected dataset: {chosen}")
    return chosen, kind

DATA_PATH, DATA_KIND = find_dataset_here()

# -------------------------
# Robust cleaners for unit-bearing text fields
# -------------------------
_num_pat = re.compile(r'[-+]?\d*\.?\d+')

def _num_from_text(x):
    if pd.isna(x): return np.nan
    if isinstance(x, (int,float,np.number)): return float(x)
    m = _num_pat.search(str(x))
    return float(m.group()) if m else np.nan

def clean_spo2(col: pd.Series) -> pd.Series:
    # "97%", "97 %", 97, 0.97 â†’ percent in [0, 100]
    out = col.apply(_num_from_text)
    out = out.where(out > 1.5, out * 100.0)  # if stored as 0.97 â†’ 97
    return pd.to_numeric(out, errors="coerce").clip(lower=0, upper=100)

def clean_temperature_c(col: pd.Series) -> pd.Series:
    # "38.5 C", "101.3 F" â†’ Â°C (assume C; convert if clearly F)
    out = col.apply(_num_from_text)
    # Heuristic: if value > 45, it's probably Fahrenheit â†’ convert to C
    out = np.where(out > 45, (out - 32.0) * 5.0/9.0, out)
    return pd.to_numeric(out, errors="coerce")

def clean_heartrate(col: pd.Series) -> pd.Series:
    # "65 bpm" â†’ 65
    return pd.to_numeric(col.apply(_num_from_text), errors="coerce")

def clean_minutes(col: pd.Series) -> pd.Series:
    # "15 min" â†’ 15
    return pd.to_numeric(col.apply(_num_from_text), errors="coerce")

def clean_ml_to_liters(col: pd.Series) -> pd.Series:
    # "600 ml" â†’ 0.6
    v = pd.to_numeric(col.apply(_num_from_text), errors="coerce")
    return v / 1000.0

def split_bp(col: pd.Series):
    # "120/80" â†’ (120, 80)
    s = col.astype(str)
    sys = pd.to_numeric(s.str.extract(r'(\d+)\s*/\s*(\d+)')[0], errors="coerce")
    dia = pd.to_numeric(s.str.extract(r'(\d+)\s*/\s*(\d+)')[1], errors="coerce")
    return sys, dia

# -------------------------
# Load + schema map + clean
# -------------------------
def _read_any(path: str, kind: str) -> pd.DataFrame:
    if kind == "csv":
        return pd.read_csv(path)
    else:
        try:
            return pd.read_excel(path, sheet_name="Sheet1")
        except Exception:
            return pd.read_excel(path)

def load_and_clean(path: str, kind: str) -> pd.DataFrame:
    raw = _read_any(path, kind)
    raw.columns = [c.strip() for c in raw.columns]

    # Map incoming headers -> canonical
    rename_map = {
        "patientId": ID_COL,
        "observationStart": TS_COL if "observationStart" in raw.columns else None,
        "observationEnd":   TS_COL if ("observationStart" not in raw.columns and "observationEnd" in raw.columns) else None,

        "stepsTaken": "steps",
        "calorieIntake": "calorie_intake",
        "sleepHours": "sleep_hours",
        "waterIntakeMl": "water_intake",
        "bathroomVisits": "bathroom_visits",

        "heartRate": "heart_rate",
        "spo2": "spo2",
        "temperature": "temperature",
        "bloodPressure": "bloodPressure",
        "mealsSkipped": "meals_skipped",
        "exerciseMinutes": "exercise_minutes",
    }
    rename_map = {k:v for k,v in rename_map.items() if v is not None}
    df = raw.rename(columns=rename_map).copy()

    # Ensure ID + TS
    if ID_COL not in df.columns:
        df[ID_COL] = "user_1"
    if TS_COL not in df.columns:
        # derive a simple daily timeline if absolutely missing
        df["_seq"] = range(len(df))
        df[TS_COL] = pd.Timestamp("2025-01-01") + pd.to_timedelta(df["_seq"], unit="D")
        df.drop(columns=["_seq"], inplace=True)

    # Parse timestamp (keep tz-aware if present)
    df[TS_COL] = pd.to_datetime(df[TS_COL], errors="coerce")
    if df[TS_COL].isna().any():
        bad = df.loc[df[TS_COL].isna(), [TS_COL]].head(3)
        raise ValueError(f"Unparseable timestamps in first rows:\n{bad}")

    # Robust vital cleaners
    if "spo2" in df.columns:
        df["spo2"] = clean_spo2(df["spo2"])
    if "temperature" in df.columns:
        df["temperature"] = clean_temperature_c(df["temperature"])
    if "heart_rate" in df.columns:
        df["heart_rate"] = clean_heartrate(df["heart_rate"])
    if "bloodPressure" in df.columns:
        sys, dia = split_bp(df["bloodPressure"])
        df["bp_sys"] = sys; df["bp_dia"] = dia
        df.drop(columns=["bloodPressure"], inplace=True, errors="ignore")
    else:
        df["bp_sys"] = df.get("bp_sys", np.nan)
        df["bp_dia"] = df.get("bp_dia", np.nan)

    # Behaviours
    if "exercise_minutes" in df.columns:
        df["exercise_minutes"] = clean_minutes(df["exercise_minutes"])
    if "water_intake" in df.columns:
        # convert mL â†’ L if stored as mL
        df["water_intake"] = clean_ml_to_liters(df["water_intake"])
    for c in ["steps","calorie_intake","sleep_hours","bathroom_visits","meals_skipped"]:
        if c in df.columns:
            df[c] = pd.to_numeric(df[c].apply(_num_from_text), errors="coerce")

    # Sort
    df[ID_COL] = df[ID_COL].fillna("user_1").astype(str)
    df = df.sort_values([ID_COL, TS_COL]).reset_index(drop=True)
    return df

df_raw = load_and_clean(DATA_PATH, DATA_KIND)

# -------------------------
# DAILY aggregation per user_id + date
# -------------------------
def to_date(ts: pd.Series) -> pd.Series:
    # Convert to naive date safely whether tz-aware or not
    ts = pd.to_datetime(ts, errors="coerce")
    if pd.api.types.is_datetime64tz_dtype(ts):
        ts = ts.dt.tz_convert(None)
    return ts.dt.date

df_raw["date"] = to_date(df_raw[TS_COL])

daily_agg = {
    # behaviours â†’ sums (or max for sleep if your data captures "max hours slept" per day)
    "steps": "sum",
    "calorie_intake": "sum",
    "water_intake": "sum",
    "bathroom_visits": "sum",
    "exercise_minutes": "sum",
    "sleep_hours": "max",

    # vitals â†’ worst/central tendency per day
    "spo2": "min",          # capture lowest oxygen
    "temperature": "max",   # capture highest fever
    "bp_sys": "max",        # worst case in-day
    "bp_dia": "max",
    "heart_rate": "mean",   # average HR; change to "max" if spikes matter more
    "meals_skipped": "sum"  # per-day total skipped
}
# Keep only columns present:
daily_agg = {k:v for k,v in daily_agg.items() if k in df_raw.columns}

daily = (df_raw
         .groupby([ID_COL,"date"], as_index=False)
         .agg(daily_agg))

# Recreate a daily timestamp at 23:59 to keep a single time per day
daily[TS_COL] = pd.to_datetime(daily["date"]) + pd.to_timedelta(23, "h") + pd.to_timedelta(59, "m")
daily = daily.drop(columns=["date"])
daily = daily.sort_values([ID_COL, TS_COL]).reset_index(drop=True)

# -------------------------
# Prepare data for models
#   - For model input, we impute missing values with column medians (keeps models happy)
#   - For reasons/risk, we use the true daily values (do NOT zero-fill vitals)
# -------------------------
model_df = daily.copy()

# Build a copy with numeric model features and median imputation
model_input = model_df.copy()
for c in FULL_FEATURES:
    if c not in model_input.columns:
        model_input[c] = np.nan
# Median impute per-column
medians = {c: model_input[c].median(skipna=True) for c in FULL_FEATURES}
for c in FULL_FEATURES:
    model_input[c] = pd.to_numeric(model_input[c], errors="coerce").fillna(medians[c])

# Scale
scaler = StandardScaler()
model_scaled = model_input.copy()
model_scaled[FULL_FEATURES] = scaler.fit_transform(model_input[FULL_FEATURES])
joblib.dump(scaler, os.path.join(OUTDIR, "scaler.pkl"))

# -------------------------
# Sequence utilities (for LSTM)
# -------------------------
def make_sequences(df_user: pd.DataFrame, seq_len: int, feat_cols: List[str]):
    X, ends = [], []
    vals = df_user[feat_cols].values.astype(float)
    for i in range(len(vals) - seq_len + 1):
        X.append(vals[i:i+seq_len]); ends.append(df_user.iloc[i+seq_len-1][TS_COL])
    return np.array(X), np.array(ends)

def stack_all_users(df_scaled: pd.DataFrame, seq_len: int, feat_cols: List[str]):
    X_all, ends_all, owners = [], [], []
    for uid, g in df_scaled.groupby(ID_COL):
        X, ends = make_sequences(g, seq_len, feat_cols)
        if len(X) == 0: continue
        X_all.append(X); ends_all.append(ends); owners += [uid]*len(X)
    if not X_all:
        raise ValueError("No sequences created. Reduce SEQ_LEN or ensure enough daily rows per user.")
    return np.vstack(X_all), np.concatenate(ends_all), np.array(owners)

class LSTMAE(nn.Module):
    def __init__(self, n_features, hidden=32, latent=16):
        super().__init__()
        self.enc = nn.LSTM(n_features, hidden, batch_first=True)
        self.enc_lin = nn.Linear(hidden, latent)
        self.dec = nn.LSTM(n_features, hidden, batch_first=True)
        self.dec_lin = nn.Linear(hidden, n_features)
    def forward(self, x):
        e,_ = self.enc(x)
        _ = self.enc_lin(e)
        d,_ = self.dec(x)
        return self.dec_lin(d)

def train_lstm(X, n_epochs=30, lr=1e-3, batch=128, device="cpu"):
    model = LSTMAE(X.shape[-1]).to(device)
    opt = torch.optim.Adam(model.parameters(), lr=lr)
    loss_fn = nn.MSELoss()
    dl = torch.utils.data.DataLoader(torch.tensor(X, dtype=torch.float32), batch_size=batch, shuffle=True)
    for ep in range(1, n_epochs+1):
        losses = []
        for xb in dl:
            xb = xb.to(device)
            pred = model(xb)
            loss = loss_fn(pred, xb)
            opt.zero_grad(); loss.backward(); opt.step()
            losses.append(loss.item())
        if ep % 5 == 0:
            print(f"[LSTM] epoch {ep:02d} loss {np.mean(losses):.5f}")
    return model

def recon_error(model, X, device="cpu"):
    model.eval()
    with torch.no_grad():
        out = model(torch.tensor(X, dtype=torch.float32).to(device)).cpu().numpy()
    return ((out - X)**2).mean(axis=(1,2))

# -------------------------
# Train anomaly engine
# -------------------------
engine_used = ENGINE if TORCH_OK or ENGINE=="iforest" else "iforest"

if engine_used == "lstm":
    X_seq, ends, owners = stack_all_users(model_scaled, SEQ_LEN, FULL_FEATURES)
    device = "cuda" if (TORCH_OK and torch.cuda.is_available()) else "cpu"
    model = train_lstm(X_seq, n_epochs=30, lr=1e-3, batch=128, device=device)
    torch.save(model.state_dict(), os.path.join(OUTDIR, "lstm.pt"))
    err = recon_error(model, X_seq, device=device)
    seq_scores = pd.DataFrame({ID_COL: owners, TS_COL: ends, "recon_error": err})
else:
    iforest = IsolationForest(n_estimators=300, contamination=0.05, random_state=RANDOM_STATE)
    iforest.fit(model_scaled[FULL_FEATURES])
    err = -iforest.score_samples(model_scaled[FULL_FEATURES])  # higher => more anomalous
    seq_scores = model_scaled[[ID_COL, TS_COL]].copy()
    seq_scores["recon_error"] = err
    joblib.dump(iforest, os.path.join(OUTDIR, "iforest.pkl"))

# Normalize error â†’ anom_score (0..1)
err_min, err_max = float(np.min(seq_scores["recon_error"])), float(np.max(seq_scores["recon_error"]))
anom_score = (seq_scores["recon_error"] - err_min) / (err_max - err_min + 1e-12)
seq_scores["anom_score"] = anom_score

# Calibrate thresholds by distribution (on raw error)
err_p80 = float(np.percentile(seq_scores["recon_error"], 80))
err_p95 = float(np.percentile(seq_scores["recon_error"], 95))
with open(os.path.join(OUTDIR, "thresholds.json"), "w") as f:
    json.dump({"err_min": err_min, "err_max": err_max, "err_p80": err_p80, "err_p95": err_p95}, f, indent=2)

# -------------------------
# Behavioural classifier (weak labels from error tail)
# -------------------------
def add_feats(df: pd.DataFrame, cols: List[str]) -> pd.DataFrame:
    g = df.sort_values([ID_COL, TS_COL]).copy()
    for f in cols:
        g[f+"_delta"]  = g.groupby(ID_COL)[f].transform(lambda s: s.diff())
        g[f+"_rmean7"] = g.groupby(ID_COL)[f].transform(lambda s: s.rolling(7, min_periods=1).mean())
        rstd = g.groupby(ID_COL)[f].transform(lambda s: s.rolling(7, min_periods=2).std())
        g[f+"_rz7"]    = (g[f] - g[f+"_rmean7"]) / rstd.replace(0, np.nan)
    return g.replace([np.inf,-np.inf], np.nan).fillna(0)

feats_df = add_feats(model_scaled[[ID_COL, TS_COL] + FULL_FEATURES].copy(), FULL_FEATURES)

tmp = feats_df[[ID_COL, TS_COL]].merge(seq_scores[[ID_COL, TS_COL, "recon_error"]],
                                       on=[ID_COL, TS_COL], how="left")
cutoff = np.percentile(tmp["recon_error"].fillna(tmp["recon_error"].median()), 95)
labels = (tmp["recon_error"] >= cutoff).astype(int)
feats_df["label"] = labels

clf_cols = []
for f in FULL_FEATURES:
    clf_cols += [f, f+"_delta", f+"_rmean7", f+"_rz7"]

X = feats_df[clf_cols].values
y = labels.values
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=TEST_SIZE, random_state=RANDOM_STATE,
    stratify=y if (y.sum()>0 and y.sum()<len(y)) else None
)

def build_clf(name: str):
    if name == "rf":
        return RandomForestClassifier(n_estimators=300, max_depth=None, min_samples_split=4,
                                      random_state=RANDOM_STATE, n_jobs=-1)
    elif name == "mlp":
        return MLPClassifier(hidden_layer_sizes=(64,32), activation="relu", max_iter=300,
                             random_state=RANDOM_STATE)
    else:
        raise ValueError("CLF must be 'rf' or 'mlp'")

clf = build_clf(CLF)
clf.fit(X_train, y_train)
if y_test.sum()>0 and y_test.sum()<len(y_test):
    print("[Classifier] eval:\n", classification_report(y_test, clf.predict(X_test), digits=3))
joblib.dump(clf, os.path.join(OUTDIR, "clf.pkl"))

# -------------------------
# Risk logic (vitals + model)  â€” only final 'risk_level' + 'reason'
# -------------------------
def vital_risk_and_reasons(row):
    reasons_med, reasons_high = [], []

    # SpO2
    s = row.get("spo2", np.nan)
    if pd.notna(s):
        if s < 90: reasons_high.append(f"SpO2 {int(round(s))}% (<90)")
        elif s < 95: reasons_med.append(f"SpO2 {int(round(s))}% (90â€“94)")

    # Temperature Â°C
    t = row.get("temperature", np.nan)
    if pd.notna(t):
        if t >= 39.4: reasons_high.append(f"High fever {t:.1f} Â°C (â‰¥39.4)")
        elif t >= 38.0: reasons_med.append(f"Fever {t:.1f} Â°C (â‰¥38.0)")

    # BP
    sys = row.get("bp_sys", np.nan); dia = row.get("bp_dia", np.nan)
    if pd.notna(sys) and pd.notna(dia):
        if sys >= 180 or dia >= 120: reasons_high.append(f"Hypertensive crisis {int(sys)}/{int(dia)}")
        elif sys >= 140 or dia >= 90: reasons_high.append(f"Stage 2 HTN {int(sys)}/{int(dia)}")
        elif (130 <= sys <= 139) or (80 <= dia <= 89): reasons_med.append(f"Stage 1 HTN {int(sys)}/{int(dia)}")

    # Exercise/day
    ex = row.get("exercise_minutes", np.nan)
    if pd.notna(ex):
        if ex < 10: reasons_high.append(f"Very low activity ({int(ex)} min)")
        elif ex < 20: reasons_med.append(f"Low activity ({int(ex)} min)")

    # Meals skipped/day
    ms = row.get("meals_skipped", np.nan)
    if pd.notna(ms):
        if ms >= 3: reasons_high.append(f"Meals skipped: {int(ms)}")
        elif ms == 2: reasons_med.append("Meals skipped: 2")

    if reasons_high: return "High", reasons_high + reasons_med
    if reasons_med:  return "Medium", reasons_med
    return "Low", []

def combine_risk(model_risk, vital_risk):
    order = {"Low":0,"Medium":1,"High":2}
    return model_risk if order[model_risk] >= order[vital_risk] else vital_risk

def model_risk_from_err_prob(recon_err, prob, p80, p95):
    if (recon_err >= p95) or (prob >= 0.85): return "High"
    if (recon_err >= p80) or (prob >= 0.65): return "Medium"
    return "Low"

# Build final table
out = feats_df[[ID_COL, TS_COL]].merge(
    seq_scores[[ID_COL, TS_COL, "recon_error", "anom_score"]],
    on=[ID_COL, TS_COL], how="left"
).sort_values([ID_COL, TS_COL])

# Classifier probability
if hasattr(clf, "predict_proba"):
    prob = clf.predict_proba(feats_df[clf_cols])[:,1]
else:
    rawp = clf.decision_function(feats_df[clf_cols])
    prob = (rawp - rawp.min()) / (rawp.max() - rawp.min() + 1e-12)
out["clf_prob"] = prob

# Attach daily vitals/behaviours for reasons (from true daily values, NOT imputed)
attach_cols = [c for c in ["spo2","temperature","bp_sys","bp_dia","exercise_minutes","meals_skipped"] if c in model_df.columns]
out = out.merge(model_df[[ID_COL, TS_COL] + attach_cols], on=[ID_COL, TS_COL], how="left")

final_levels, reasons = [], []
for _, row in out.iterrows():
    m_risk = model_risk_from_err_prob(row["recon_error"], row["clf_prob"], err_p80, err_p95)
    v_risk, v_flags = vital_risk_and_reasons(row)
    final = combine_risk(m_risk, v_risk)
    final_levels.append(final)

    rs = []
    if final in ("Medium","High"):
        if m_risk == "High":
            if row["recon_error"] >= err_p95: rs.append("Strong sequence anomaly (â‰¥95th percentile)")
            if row["clf_prob"] >= 0.85: rs.append("Classifier: strong behavioural anomaly (â‰¥0.85)")
        elif m_risk == "Medium":
            if row["recon_error"] >= err_p80: rs.append("Moderate sequence anomaly (â‰¥80th percentile)")
            if 0.65 <= row["clf_prob"] < 0.85: rs.append("Classifier: possible behavioural anomaly (â‰¥0.65)")
        rs += v_flags
    reasons.append("; ".join(rs))

out["risk_level"] = final_levels
out["reason"] = reasons

alerts = out[[ID_COL, TS_COL, "anom_score", "clf_prob", "risk_level", "reason"]].copy()
alerts_path = os.path.join(OUTDIR, "alerts.csv")
alerts.to_csv(alerts_path, index=False)
print(f"[done] Saved alerts â†’ {alerts_path}  (rows={len(alerts)})")

# -------------------------
# Visualizations
# -------------------------
plt.figure(figsize=(7,4))
plt.hist(alerts["anom_score"].dropna(), bins=40)
# map raw error thresholds into 0..1 for reference
def norm_err_to_anom(e, lo, hi):
    return 0.0 if math.isclose(hi, lo) else (e - lo) / (hi - lo)
anom_p80 = norm_err_to_anom(err_p80, err_min, err_max)
anom_p95 = norm_err_to_anom(err_p95, err_min, err_max)
plt.axvline(anom_p80, linestyle="--", linewidth=1.5)
plt.axvline(anom_p95, linestyle="-.", linewidth=1.5)
plt.title("Anomaly Score Distribution")
plt.xlabel("anom_score (0â€“1)"); plt.ylabel("count")
plt.show()

plt.figure(figsize=(7,5))
for lvl in ["Low","Medium","High"]:
    sel = alerts[alerts["risk_level"]==lvl]
    plt.scatter(sel["anom_score"], sel["clf_prob"], s=18, alpha=0.7, label=lvl)
plt.legend(); plt.xlabel("anom_score"); plt.ylabel("clf_prob")
plt.title("Model Signals vs Final Risk"); plt.show()

plt.figure(figsize=(5,3.8))
alerts["risk_level"].value_counts().reindex(["Low","Medium","High"]).plot(kind="bar")
plt.title("Risk Level Counts"); plt.xlabel("risk_level"); plt.ylabel("count"); plt.xticks(rotation=0)
plt.show()

# Example timeline for the patient with highest risk
top_user = (alerts.assign(rank=alerts["risk_level"].map({"Low":0,"Medium":1,"High":2}))
                   .sort_values(["rank", TS_COL], ascending=[False, True])[ID_COL].iloc[0])
u = model_df[model_df[ID_COL]==top_user].merge(alerts[[ID_COL, TS_COL, "risk_level"]],
                                               on=[ID_COL, TS_COL], how="left")
fig, axes = plt.subplots(4, 1, figsize=(10,10), sharex=True)
if "steps" in u: axes[0].plot(u[TS_COL], u["steps"]); axes[0].set_ylabel("steps")
if "sleep_hours" in u: axes[1].plot(u[TS_COL], u["sleep_hours"]); axes[1].set_ylabel("sleep_hours")
if "water_intake" in u: axes[2].plot(u[TS_COL], u["water_intake"]); axes[2].set_ylabel("water_L")
if "spo2" in u: axes[3].plot(u[TS_COL], u["spo2"]); axes[3].set_ylabel("SpO2")
axes[-1].set_xlabel("date")
fig.suptitle(f"Daily timeline â€“ user {top_user}"); plt.show()

alerts.head(10)
```
