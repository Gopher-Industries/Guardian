```python
# --- Validator for updated GM Alerts dataset (AI spreadsheet schema) ---
# Works with:
#   - GmAlertsDataset.csv  (recommended, AI-aligned)
#   - GMAlertsDataset.csv  (fallback)
# Accepts either AI schema (heartRate/spo2/temperature/bloodPressure)
# or old dotted schema (vitals.* / adls.*) automatically.

import os, re
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display

# 1) Load (try common filenames)
cand_paths = ["New AI spreadsheet - Sheet1.csv", "GMAlertsDataset.csv"]
path = next((p for p in cand_paths if os.path.exists(p)), cand_paths[0])
df = pd.read_csv(path)
print(f"Loaded: {path}  (rows={len(df)})")

# 2) Column helpers (support both AI and old dotted schema)
def pick_col(*names, default=None):
    for n in names:
        if n in df.columns:
            return n
    return default

COL_SPO2 = pick_col("spo2", "vitals.spo2.value")
COL_TEMP = pick_col("temperature", "vitals.temperature.value")
COL_BP   = pick_col("bloodPressure")  # AI schema
COL_SYS  = pick_col("vitals.bloodPressure.systolic")
COL_DIA  = pick_col("vitals.bloodPressure.diastolic")
COL_EX   = pick_col("exerciseMinutes", "adls.exerciseMinutes")
COL_MEAL = pick_col("mealsSkipped", "adls.mealsSkipped")

# 3) Robust parsing to numeric
def to_num_series(s):
    if s is None or s not in df.columns:
        return pd.Series([np.nan]*len(df))
    x = df[s].astype(str)
    # strip units like "bpm", "%", "Â°C", "mmHg"
    x = x.str.replace(r"[^\d\.\-]+", "", regex=True)
    return pd.to_numeric(x, errors="coerce")

# Blood pressure parsing
def parse_bp_pair():
    """Return (sys, dia) as numeric from either 'bloodPressure' or dotted cols."""
    if COL_BP and COL_BP in df.columns:
        bp = df[COL_BP].astype(str)
        # patterns like "121/70" or "121/70 mmHg"
        m = bp.str.extract(r"(?P<sys>\d+(?:\.\d+)?)\s*/\s*(?P<dia>\d+(?:\.\d+)?)")
        sys = pd.to_numeric(m["sys"], errors="coerce")
        dia = pd.to_numeric(m["dia"], errors="coerce")
        return sys, dia
    # dotted fallback
    sys = to_num_series(COL_SYS)
    dia = to_num_series(COL_DIA)
    return sys, dia

spo2 = to_num_series(COL_SPO2)
temp = to_num_series(COL_TEMP)
bp_sys, bp_dia = parse_bp_pair()
ex   = to_num_series(COL_EX)
meals= to_num_series(COL_MEAL)

# 4) Threshold helpers (exactly as specified)
def spo2_risk(v):
    if pd.isna(v): return "Low"
    if v >= 95: return "Low"
    if 90 <= v <= 94: return "Medium"
    return "High"

def temp_risk(v):
    if pd.isna(v): return "Low"
    if v < 38.0: return "Low"
    if 38.0 <= v <= 39.3: return "Medium"
    return "High"

def bp_risk(sys, dia):
    sys = np.nan if pd.isna(sys) else sys
    dia = np.nan if pd.isna(dia) else dia
    if pd.isna(sys) or pd.isna(dia): return "Low"
    if sys >= 140 or dia >= 90: return "High"          # Stage 2 / crisis handled here
    if (130 <= sys <= 139) or (80 <= dia <= 89): return "Medium"
    return "Low"

def exercise_risk(m):
    if pd.isna(m): return "Low"
    if m >= 20: return "Low"
    if 10 <= m <= 19: return "Medium"
    return "High"

def meals_risk(k):
    if pd.isna(k): return "Low"
    if k <= 1: return "Low"
    if k == 2: return "Medium"
    return "High"

# 5) Compute per-feature & overall risk (max severity)
df["_risk_spo2"]  = [spo2_risk(v) for v in spo2]
df["_risk_temp"]  = [temp_risk(v) for v in temp]
df["_risk_bp"]    = [bp_risk(s, d) for s, d in zip(bp_sys, bp_dia)]
df["_risk_ex"]    = [exercise_risk(v) for v in ex]
df["_risk_meals"] = [meals_risk(v) for v in meals]

order = {"Low": 0, "Medium": 1, "High": 2}
rev   = {0: "Low", 1: "Medium", 2: "High"}
row_max = np.maximum.reduce([
    pd.Series(df["_risk_spo2"]).map(order).fillna(0).values,
    pd.Series(df["_risk_temp"]).map(order).fillna(0).values,
    pd.Series(df["_risk_bp"]).map(order).fillna(0).values,
    pd.Series(df["_risk_ex"]).map(order).fillna(0).values,
    pd.Series(df["_risk_meals"]).map(order).fillna(0).values,
])
df["_risk_overall_recomputed"] = pd.Series(row_max).map(rev)

# 6) Checks & proportions
has_declared = "risk_category" in df.columns
if has_declared:
    print("== Declared dataset proportion ==")
    print((df["risk_category"].value_counts(normalize=True)*100).round(2))
else:
    print("== Declared dataset proportion ==\n(no 'risk_category' column in file)")

print("\n== Recomputed (from vitals) proportion ==")
print((df["_risk_overall_recomputed"].value_counts(normalize=True)*100).round(2))

if has_declared:
    match_rate = (df["_risk_overall_recomputed"] == df["risk_category"]).mean() * 100
    print(f"\nExact match between declared risk_category and recomputed overall risk: {match_rate:.2f}%")

# 7) Per-feature agreement with overall (sanity)
for col, lab in [
    ("_risk_spo2",  "SpOâ‚‚"),
    ("_risk_temp",  "Temperature"),
    ("_risk_bp",    "Blood Pressure"),
    ("_risk_ex",    "Exercise minutes"),
    ("_risk_meals", "Meals skipped")
]:
    agree = (df[col] == df["_risk_overall_recomputed"]).mean() * 100
    print(f"{lab:>18} agrees with OVERALL risk: {agree:6.2f}%")

# 8) Visuals by risk (uses numeric columns derived above)
def plot_hist_by_risk(values, title, bins=20):
    for risk in ["Low", "Medium", "High"]:
        plt.figure(figsize=(6,4))
        subset = values[df["_risk_overall_recomputed"] == risk]
        plt.hist(pd.to_numeric(subset, errors="coerce").dropna(), bins=bins)
        plt.title(f"{title} â€” {risk} group")
        plt.xlabel(title); plt.ylabel("Count")
        plt.tight_layout(); plt.show()

plot_hist_by_risk(spo2,         "SpOâ‚‚ (%)")
plot_hist_by_risk(temp,         "Temperature (Â°C)")
plot_hist_by_risk(bp_sys,       "Systolic (mmHg)")
plot_hist_by_risk(bp_dia,       "Diastolic (mmHg)")
plot_hist_by_risk(ex,           "Exercise minutes/day")
plot_hist_by_risk(meals,        "Meals skipped/day")

# 9) Quick cross-tabs
def show_crosstab(values, label):
    s = pd.to_numeric(values, errors="coerce")
    ct = pd.crosstab(df["_risk_overall_recomputed"], pd.cut(s, bins=10), normalize="index").round(3)
    print(f"\nCrosstab for {label} (by recomputed overall risk)")
    display(ct)

show_crosstab(spo2,   "SpOâ‚‚")
show_crosstab(temp,   "Temperature")
show_crosstab(bp_sys, "Systolic BP")
show_crosstab(bp_dia, "Diastolic BP")
show_crosstab(ex,     "Exercise minutes")
show_crosstab(meals,  "Meals skipped")
```
