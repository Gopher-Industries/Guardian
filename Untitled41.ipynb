```python
import re
import ast
import pandas as pd
from transformers import pipeline

# Load HuggingFace biomedical NER
try:
    ner_pipeline = pipeline(
        "ner",
        model="d4data/biomedical-ner-all",
        aggregation_strategy="simple"
    )
    print(" Biomedical NER loaded")
except Exception as e:
    ner_pipeline = None
    print("Biomedical NER not available:", e)
```

```python
def extract_medications(text):
    meds = set()

    # HuggingFace Biomedical NER
    if ner_pipeline:
        try:
            ents = ner_pipeline(text)
            for ent in ents:
                if ent["entity_group"].lower() in ["drug", "chemical"]:
                    meds.add(ent["word"])
        except Exception as e:
            print("NER failed:", e)

    # Regex + suffix fallback
    med_pattern = r"\b([A-Z][a-z]+(?:pril|olol|sartan|dipine|azole|mycin|cillin|mab|vir|phen|done|xone|zole|dine|mine|mide))\b"
    fallback = re.findall(med_pattern, text)
    meds.update(fallback)

    return list(set(meds))
```

```python
def extract_entities_rule_based(text):
    entities = {"symptoms": [], "vitals": [], "procedures": [], "medications": []}
    parsed_vitals = {}

    if not isinstance(text, str) or not text.strip():
        return entities

    # --- Symptoms ---
    symptoms_keywords = ["pain", "fatigue", "shortness of breath", "tired", "uncomfortable", "sleepy","ill"]
    for kw in symptoms_keywords:
        if kw in text.lower():
            entities["symptoms"].append(kw)

    # --- Vitals ---
    vitals_patterns = {
        "heart_rate": r"heart rate\s*(of|is|was)?\s*(\d+)",
        "spo2": r"oxygen (saturation)?\s*(at|of|is|was|remained)?\s*(\d+)%",
        "temperature": r"temperature\s*(is|was)?\s*(\d+(\.\d+)?)\s*°?C",
        "bp": r"blood pressure\s*(of|is|was)?\s*(\d{2,3})/(\d{2,3})"
    }

    if m := re.search(vitals_patterns["heart_rate"], text.lower()):
        val = int(m.group(2))
        entities["vitals"].append(f"heart rate of {val} bpm")
        parsed_vitals["heart_rate"] = val

    if m := re.search(vitals_patterns["spo2"], text.lower()):
        val = int(m.group(3))
        entities["vitals"].append(f"oxygen saturation at {val}%")
        parsed_vitals["spo2"] = val

    if m := re.search(vitals_patterns["temperature"], text.lower()):
        val = float(m.group(2))
        entities["vitals"].append(f"temperature is {val}°C")
        parsed_vitals["temperature"] = val

    if m := re.search(vitals_patterns["bp"], text.lower()):
        sys, dia = int(m.group(2)), int(m.group(3))
        entities["vitals"].append(f"blood pressure of {sys}/{dia} mmHg")
        parsed_vitals["bp_sys"], parsed_vitals["bp_dia"] = sys, dia

    # --- Procedures ---
    procedures = ["xray", "ecg", "mri", "ct scan"]
    for p in procedures:
        if p in text.lower():
            entities["procedures"].append(p)

    # --- Medications ---
    entities["medications"] = extract_medications(text)

    entities["parsed_vitals"] = parsed_vitals
    return entities
```

```python
def generate_alerts(entities):
    alerts = {"guardian": [], "nurse": []}

    if not entities or not isinstance(entities, dict):
        return alerts

    # --- Vitals Alerts ---
    vitals = entities.get("parsed_vitals", {})

    if "heart_rate" in vitals:
        hr = vitals["heart_rate"]
        if hr > 120:
            alerts["nurse"].append(f"High heart rate detected: {hr} bpm")
        elif hr < 50:
            alerts["nurse"].append(f"Low heart rate detected: {hr} bpm")

    if "spo2" in vitals:
        spo2 = vitals["spo2"]
        if spo2 < 92:
            alerts["nurse"].append(f"Low oxygen saturation: {spo2}%")
            alerts["guardian"].append("Patient may be experiencing breathing issues")

    if "temperature" in vitals:
        temp = vitals["temperature"]
        if temp > 38.0:
            alerts["nurse"].append(f"Fever detected: {temp}°C")
        elif temp < 35.0:
            alerts["nurse"].append(f"Hypothermia risk: {temp}°C")

    if "bp_sys" in vitals and "bp_dia" in vitals:
        sys, dia = vitals["bp_sys"], vitals["bp_dia"]
        if sys > 160 or dia > 100:
            alerts["nurse"].append(f"High blood pressure detected: {sys}/{dia} mmHg")
        elif sys < 90 or dia < 60:
            alerts["nurse"].append(f"Low blood pressure detected: {sys}/{dia} mmHg")

    # --- Symptoms Alerts ---
    for s in entities.get("symptoms", []):
        if s in ["pain", "fatigue", "shortness of breath", "uncomfortable"]:
            alerts["guardian"].append(f"Patient reports: {s}")
            alerts["nurse"].append(f"Monitor patient for: {s}")

    # --- Medications Alerts ---
    for m in entities.get("medications", []):
        if m.lower() in ["furosemide", "lasix"]:  # diuretic
            alerts["nurse"].append("Monitor fluid balance: Furosemide administered")
        if m.lower() in ["insulin"]:  # diabetes med
            alerts["nurse"].append("Monitor blood glucose after insulin dose")

    return alerts
```

```python
# Apply alerts for each row
df["alerts"] = df["entitiesExtracted_new"].apply(generate_alerts)

# Show preview
df[[text_column, "entitiesExtracted_new", "alerts"]].head(5)
```

```python

def extract_entities_rule_based(text):
    entities = {"symptoms": [], "vitals": [], "procedures": [], "medications": [], "parsed_vitals": {}}
    if not text or not isinstance(text, str):
        return entities

    # Run NER model
    try:
        ner_results = ner_pipeline(text)
    except Exception:
        ner_results = []

    # Add biomedical entities
    for ent in ner_results:
        label = ent["entity_group"].lower()
        word = ent["word"].strip()
        if label in ["drug", "medication"]:
            entities["medications"].append(word.capitalize())
        elif label in ["symptom", "disease"]:
            entities["symptoms"].append(word.lower())
        elif label in ["procedure", "test"]:
            entities["procedures"].append(word.lower())

    # Regex for vitals
    vitals_patterns = {
        "heart_rate": r"(heart rate (is|of)?\s*\d+\s*bpm)",
        "spo2": r"(oxygen (saturation)?( remained| is| at)?\s*\d+%)",
        "bp": r"(blood pressure (of|is)?\s*\d{2,3}/\d{2,3})",
        "temperature": r"(temperature (is|was)?\s*\d+(\.\d+)?\s*(°C|C|F)?)",
        "oxygen_flow": r"(oxygen (at|via)\s*\d+L/min)"
    }

    for key, pattern in vitals_patterns.items():
        matches = re.findall(pattern, text, flags=re.IGNORECASE)
        for m in matches:
            val = m[0] if isinstance(m, tuple) else m
            entities["vitals"].append(val.strip())

    # Parse vitals into numbers for alerts
    hr = re.search(r"heart rate (is|of)?\s*(\d+)", text, flags=re.IGNORECASE)
    spo2 = re.search(r"oxygen (saturation)?( remained| is| at)?\s*(\d+)%", text, flags=re.IGNORECASE)
    bp = re.search(r"blood pressure (of|is)?\s*(\d{2,3})/(\d{2,3})", text, flags=re.IGNORECASE)

    if hr: entities["parsed_vitals"]["heart_rate"] = int(hr.group(2))
    if spo2: entities["parsed_vitals"]["spo2"] = int(spo2.group(3))
    if bp:
        entities["parsed_vitals"]["bp_sys"] = int(bp.group(2))
        entities["parsed_vitals"]["bp_dia"] = int(bp.group(3))

    return entities
```

```python
import re
import json
import pandas as pd
from transformers import pipeline
```

```python
# Try loading biomedical NER model
try:
    ner_pipeline = pipeline("ner", model="d4data/biomedical-ner-all", aggregation_strategy="simple")
    print("Biomedical NER loaded successfully ")
except Exception as e:
    ner_pipeline = None
    print("Biomedical NER not available (falling back to regex-only meds). Error:", e)
```

```python
def clean_medications(meds):
    """Merge subword tokens like 'Am', '##lo', '##dipine' into 'Amlodipine'."""
    cleaned = []
    current = ""
    for m in meds:
        if m.startswith("##"):
            current += m[2:]
        else:
            if current:
                cleaned.append(current.capitalize())
            current = m
    if current:
        cleaned.append(current.capitalize())
    return list(set(cleaned))  # remove duplicates
```

```python
# Regex patterns for vitals
vital_patterns = {
    "heart_rate": r"heart rate\s*(of|is|was)?\s*(\d+)\s*bpm",
    "spo2": r"(oxygen saturation\s*(at|of|is|was)?\s*(\d+)%)",
    "blood_pressure": r"blood pressure\s*(of|is|was)?\s*(\d+/\d+)",
    "temperature": r"temperature\s*(was|is)?\s*(\d+(\.\d+)?)\s*(°C|C|F)"
}

# Simple symptom keywords
symptom_keywords = ["pain", "fatigue", "shortness of breath", "tired", "uncomfortable", "sleepy","ill"]
```

```python
def extract_entities_rule_based(note):
    entities = {"symptoms": [], "vitals": [], "procedures": [], "medications": [], "parsed_vitals": {}}

    if not isinstance(note, str) or note.strip() == "":
        return entities

    # ------------------------
    # 1. Vitals extraction
    # ------------------------
    for vital, pattern in vital_patterns.items():
        match = re.search(pattern, note, re.IGNORECASE)
        if match:
            entities["vitals"].append(match.group().lower())
            if vital == "heart_rate":
                entities["parsed_vitals"]["heart_rate"] = int(match.group(2))
            elif vital == "spo2":
                entities["parsed_vitals"]["spo2"] = int(re.findall(r"\d+", match.group(0))[0])
            elif vital == "blood_pressure":
                bp = match.group(2).split("/")
                entities["parsed_vitals"]["bp_sys"] = int(bp[0])
                entities["parsed_vitals"]["bp_dia"] = int(bp[1])
            elif vital == "temperature":
                entities["parsed_vitals"]["temperature"] = float(match.group(2))

    # ------------------------
    # 2. Symptom extraction
    # ------------------------
    for symptom in symptom_keywords:
        if re.search(rf"\b{symptom}\b", note, re.IGNORECASE):
            entities["symptoms"].append(symptom)

    # ------------------------
    # 3. Medication extraction
    # ------------------------
    meds_found = []
    if ner_pipeline:  # HuggingFace NER
        ner_results = ner_pipeline(note)
        for ent in ner_results:
            if ent["entity_group"] == "Chemical":
                meds_found.append(ent["word"])

    # Clean and deduplicate medications
    if meds_found:
        entities["medications"] = clean_medications(meds_found)

    return entities
```

```python
# Load dataset (update path to your CSV file)
df = pd.read_csv(r"C:\Users\harsh\Downloads\New AI spreadsheet - Sheet1.csv")

# Pick nursingNote column (adjust if different)
if "nursingNote" in df.columns:
    text_column = "nursingNote"
elif "clinicalSummary" in df.columns:
    text_column = "clinicalSummary"
else:
    raise KeyError("No nursingNote or clinicalSummary column found in dataset.")

# Extract entities for entire dataset
df["entities_json"] = df[text_column].apply(lambda x: json.dumps(extract_entities_rule_based(x), indent=4))

# Save results
df.to_csv("dataset_with_entities.json.csv", index=False)

# Show a preview
print(df[["nursingNote", "entities_json"]].head(3))
```
